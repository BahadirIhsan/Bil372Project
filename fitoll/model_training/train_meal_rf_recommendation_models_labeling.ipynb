```python
import pandas as pd

csv_path = "../data/Food_and_Nutrition new.csv"  # kendi yolunu kullan
df = pd.read_csv(csv_path)

df.head()
```

```python
df.columns
```

```python
meal_cols = ["Breakfast Suggestion", "Lunch Suggestion", "Dinner Suggestion","Snack Suggestion"]

for col in meal_cols:
    print(f"\n==== {col} ====")
    print("Toplam satır sayısı:", df[col].shape[0])
    print("Farklı yemek sayısı  :", df[col].nunique())
```

```python
for col in meal_cols:
    print(f"\n\n######################### {col} #########################")
    vc = df[col].value_counts()
    print(vc)
```

```python
for col in meal_cols:
    print(f"\n\n######################### {col} (top 30) #########################")
    vc = df[col].value_counts()
    print(vc.head(30))
```

```python
for col in meal_cols:
    # Value counts al
    vc = df[col].value_counts().reset_index()
    vc.columns = [col, "Count"]  # sütun isimlerini düzelt
    
    # CSV olarak kaydet
    output_name = f"unique_{col.replace(' ', '_').lower()}_with_counts.csv"
    vc.to_csv(output_name, index=False)
    
    print(f"{col} için kaydedildi → {output_name}")
```
