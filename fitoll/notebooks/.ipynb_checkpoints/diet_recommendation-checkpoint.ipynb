```python
import os, numpy as np, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
save_dir = "../diet_recommendation_analysis_outputs"
os.makedirs(save_dir, exist_ok=True)
```

```python
df = pd.read_csv('../data/Food_and_Nutrition new.csv')
```

```python
#bu datada bmi yok ama hesaplayÄ±p biz koyalÄ±m iÅŸimize yarayacaktÄ±r ilerde 
```

```python
df['Height_m'] = df['Height'] / 100.0

df['BMI'] = (df['Weight'] / (df['Height_m']**2)).round(2)

def bmi_cat(bmi):
    if bmi < 18.5:   return 'Underweight'
    elif bmi < 25:   return 'Normal'
    elif bmi < 30:   return 'Overweight'
    else:            return 'Obese'

df['BMI_Class'] = df['BMI'].apply(bmi_cat)
```

```python
df['BMI'].describe()
```

```python
df['BMI_Class'].value_counts()
```

```python
#makrolardan hesaplanan toplam enerji
df['Macro_kcal'] = (df['Protein']*4 + df['Carbohydrates']*4 + df['Fat']*9)

#hedefe gÃ¶re ne kadar sapmÄ±ÅŸÄ±z (pozitif = hedefin Ã¼stÃ¼nde)
df['Calorie_Gap'] = (df['Macro_kcal'] - df['Daily Calorie Target']).round(1)

#makro yÃ¼zdeleri (toplam makro kcala gÃ¶re)
eps = 1e-6
df['Protein_%'] = (df['Protein']*4 / (df['Macro_kcal']+eps) * 100).round(1)
df['Carb_%']    = (df['Carbohydrates']*4 / (df['Macro_kcal']+eps) * 100).round(1)
df['Fat_%']     = (df['Fat']*9 / (df['Macro_kcal']+eps) * 100).round(1)

#100 kcal baÅŸÄ±na besin yoÄŸunluÄŸu (protein ve lif -> kalite sinyali)
df['Protein_per_100kcal'] = (df['Protein'] / ((df['Macro_kcal']+eps)/100)).round(2)
df['Fiber_per_100kcal']   = (df['Fiber']   / ((df['Macro_kcal']+eps)/100)).round(2)

#yaÅŸ grubu (bazÄ± modellerde sÄ±nÄ±flandÄ±rÄ±cÄ± olarak iyi Ã§alÄ±ÅŸabilir deneyeceÄŸiz)
df['Age_Group'] = pd.cut(
    df['Ages'],
    bins=[0, 18, 25, 35, 45, 60, 150],
    labels=['<=18','19-25','26-35','36-45','46-60','60+'],
    right=True
)
```

```python
#genel olarak veri dengeli ve gÃ¼zel daÄŸÄ±lmÄ±ÅŸ yani ÅŸimdi grafik analizlere geÃ§elim
```

```python
num_cols = [
    'Ages', 'Height', 'Weight', 'BMI', 'Daily Calorie Target',
    'Calories', 'Macro_kcal', 'Calorie_Gap',
    'Protein', 'Carbohydrates', 'Fat', 'Fiber',
    'Protein_%', 'Carb_%', 'Fat_%'
]#yine ÅŸÃ¶yle sayÄ±sal olanlarÄ± alalÄ±m bi
```

```python
corr = df[num_cols].corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("SayÄ±sal Ã–zellikler ArasÄ±ndaki Korelasyon IsÄ± HaritasÄ±")
plt.tight_layout()
plt.savefig(os.path.join(save_dir, "correlation_heatmap.png"), dpi=150)
plt.close()
```

```python
#korelasyon sonucumuz beklenen gibi Ã§Ä±ktÄ± birbiriyle baÄŸlantÄ±lÄ± olanlar gÃ¼Ã§lÃ¼ korelasyondayken birbirine zÄ±t olanlar
#neg korelasyonda gayet gÃ¼zel
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['BMI'], bins=30, kde=True, color='steelblue')
plt.title("BMI DeÄŸerlerinin DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("BMI (VÃ¼cut Kitle Ä°ndeksi)")
plt.ylabel("Frekans")
plt.tight_layout()
plt.savefig(os.path.join(save_dir, "bmi_distribution.png"), dpi=150)
plt.close()
```

```python
#veri setindeki Ã§oÄŸu kiÅŸi normal aralÄ±kta ama mean>median olcak cÃ¼nkÃ¼ grafik saÄŸa doÄŸru Ã§arpÄ±k
#obez kiÅŸi sayÄ±sÄ± az 
```

```python
#boxplot calori gap
plt.figure(figsize=(10,6))
sns.boxplot(
    x='Age_Group',
    y='Calorie_Gap',
    data=df,
    hue='Age_Group',           # <--- yeni ekledik
    legend=False,              # <--- uyarÄ±yÄ± susturur
    palette='cool'
)
plt.title("YaÅŸ GruplarÄ±na GÃ¶re Hedeften Sapma (Calorie Gap)")
plt.xlabel("YaÅŸ Grubu")
plt.ylabel("Calorie Gap (kcal)")
plt.tight_layout()
plt.savefig(os.path.join(save_dir, "agegroup_caloriegap_boxplot.png"), dpi=150)
plt.close()
```

```python
#ÅŸimdi burda box plotta Ã§ok fazla outlier var ama bu durum ÅŸununla aÃ§Ä±klanabilir mesela kilolu birisi kilo vermek istiyor
#o yÃ¼zden kalori gap olmasÄ± lazÄ±m  veya tam tersi almak isteyenler + outlierda olabilir bu normal bir durum beklenen gibi yani
```

```python
#bmi calori gap
bmi_gap = df.groupby('BMI_Class')['Calorie_Gap'].mean().sort_index()

plt.figure(figsize=(7,5))
bmi_gap.plot(kind='bar', color='coral', edgecolor='black')
plt.title("BMI SÄ±nÄ±fÄ±na GÃ¶re Ortalama Calorie Gap")
plt.xlabel("BMI Class")
plt.ylabel("Ortalama Calorie Gap (kcal)")
plt.tight_layout()
plt.savefig(os.path.join(save_dir, "bmi_class_caloriegap_bar.png"), dpi=150)
plt.close()
```

```python
#burda da nerdeyse her bmi grup - Ã§Ä±ktÄ± bu da Ã§oÄŸunun kilo vermek istemesiyle alakalÄ± olabilir  mesela underweight olanlarÄ±n
#calori gapi en az bu da beklenen gibi 
```

```python
#protein yÃ¼zdelerine bakalÄ±m
plt.figure(figsize=(8,5))
sns.histplot(df['Protein_%'], bins=30, kde=True, color='seagreen')
plt.title("Protein YÃ¼zdesi DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Protein (%)")
plt.ylabel("Frekans")
plt.tight_layout()
plt.savefig(os.path.join(save_dir, "protein_percentage_distribution.png"), dpi=150)
plt.close()
```

```python
#bakalÄ±m bakalÄ±m hepsi nasÄ±l daÄŸÄ±lÄ±yor 3 Ã¼ iÃ§in de tekrar oluÅŸturuyorum o yÃ¼zden acaba Ã¼Ã§Ã¼ de standart mÄ±
for col in ['Protein_%', 'Carb_%', 'Fat_%']:
    plt.figure(figsize=(8,5))
    sns.histplot(df[col], bins=30, kde=True, color='green', alpha=0.6)
    plt.title(f'{col} DaÄŸÄ±lÄ±mÄ±', fontsize=13)
    plt.xlabel('%')
    plt.ylabel('Frekans')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    # dosya adÄ± oluÅŸturma
    file_name = f"{col.replace('%','percent')}_distribution.png"
    plt.savefig(os.path.join(save_dir, file_name), dpi=150, bbox_inches='tight')
    plt.close()
```

```python
#veride gerÃ§eÄŸe uygun ÅŸekilde 30 protein 40 carb 30 fat oranlarÄ± mevcut burdan bunu anlÄ±yoruz
```

```python
#yaÅŸa gÃ¶re boxplot
macro_melted = df.melt(
    id_vars=['Age_Group'],
    value_vars=['Protein_%', 'Carb_%', 'Fat_%'],
    var_name='Macro_Type',
    value_name='Percent'
)

# 2ï¸âƒ£ Grafik Ã§iz
plt.figure(figsize=(10,6))
sns.boxplot(
    x='Age_Group',
    y='Percent',
    hue='Macro_Type',
    data=macro_melted,
    palette='Set2'
)
plt.title("YaÅŸ GruplarÄ±na GÃ¶re Makro DaÄŸÄ±lÄ±mÄ± (%)", fontsize=13)
plt.xlabel("YaÅŸ Grubu")
plt.ylabel("Makro OranÄ± (%)")
plt.legend(title="Makro TÃ¼rÃ¼", loc='upper right')
plt.tight_layout()
plt.savefig(os.path.join(save_dir, "agegroup_macro_distribution.png"), dpi=150, bbox_inches='tight')
plt.close()
```

```python
#yaÅŸa gÃ¶re bakÄ±nca da gayet dengeli bir daÄŸÄ±lÄ±m gÃ¶rÃ¼yoruz bazÄ± outlierlar var tabiki bunlar muhtemlen kilo alma
#verme isteÄŸindeki diyetlere gÃ¶re yani beklenen birÅŸey
```
