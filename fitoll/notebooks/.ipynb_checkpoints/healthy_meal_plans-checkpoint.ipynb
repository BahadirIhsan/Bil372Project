```python
outputdir='../healty_meal_analysis_outputs'
```

```python
df = pd.read_csv('../data/healthy_meal_plans.csv')
```

```python
#burda kolonları ayırırsak daha verimli olabilir cünkü bazı kolonlar sadece 0-1 gibi flag alıyor bazılarıysa direkt olarak sayısal değerler alıyor
#bu kolonları ayırcağım o yüzden 
```

```python
#is_healthy target oldugu icin onu almadım tamamı bu
num_cols_all = [c for c in df.select_dtypes(include=[np.number]).columns if c != "is_healthy"]
#flag kolonlar
binary_cols = [c for c in num_cols_all if set(df[c].unique()) <= {0,1}]
#sayısal kolonlar
cont_cols = [c for c in num_cols_all if c not in binary_cols]
```

```python
num_cols_all
```

```python
binary_cols
```

```python
cont_cols
```

```python
#sayısal olanlar için histogram yapabiliriz bununla başlayalım
for c in cont_cols:
    plt.figure(figsize=(7,4))
    plt.hist(df[c].dropna(), bins=30)
    plt.title(f"{c} - histogram")
    plt.xlabel(c); plt.ylabel("count")
    plt.tight_layout()
    plt.savefig(os.path.join(outputdir, f"hist_{c}.png"), bbox_inches="tight", dpi=300)
    plt.close()
```

```python
#tamamı genel olarak yaklaşık ve benzer dağılımda prep timeda ciddi farklar var ama bu zamansal olduğu için açıklanabilcek bir durum
```

```python
#boxplot yapalım bir de sayısallar için
for c in cont_cols:
    plt.figure(figsize=(7,2.6))
    plt.boxplot(df[c].dropna(), vert=False, widths=0.6)
    plt.title(f"{c} - boxplot")
    plt.xlabel(c)
    plt.tight_layout()
    plt.savefig(os.path.join(outputdir, f"box_{c}.png"), bbox_inches="tight", dpi=300)
    plt.close()
```

```python
#veri boxplotta da gayet dengeli ve temiz olarak görünüyor
```

```python
#simdi binary verilere geçelim
if binary_cols:
    counts = df[binary_cols].sum().sort_values(ascending=False)
    plt.figure(figsize=(9, 0.5*len(binary_cols)+2))
    plt.barh(counts.index, counts.values)
    plt.title("Diyet Etiketleri – Frekans")
    plt.xlabel("adet")
    plt.tight_layout()
    plt.savefig(os.path.join(outputdir, "diet_flags_counts.png"), bbox_inches="tight", dpi=300)
    plt.close()
```

```python
#burda diyet tiplerinin sayısal grafiği oluyor vegetarian en sık 
```

```python
# labellı bir veri zaten şimdi de is_healthy dağılımına bakalım
plt.figure(figsize=(5,3))
vals = df["is_healthy"].value_counts().sort_index()
plt.bar(["not_healthy(0)","healthy(1)"], vals.values)
plt.title("is_healthy dağılımı"); plt.ylabel("adet")
plt.tight_layout()
plt.savefig(os.path.join(outputdir, "is_healthy_distribution.png"), bbox_inches="tight", dpi=300)
plt.close()
```

```python
#çok ciddi bir şekilde not healthy dağılımı var 450-50 gibi bir şekilde yani healthy olanlar ciddi azınlıkta
#eğerki ilerde bu kısma dayalı bir sınıflandırma modeli kurcaksak yapmamız gereken şeyler var cunku burda baya imbalanced bir veri var
#ya kendimiz SMOTE ve RandomOverSampler gibi yöntemlerle is_healthy=1 olan örnekler ekleyeceğiz sete 
#ya da bazı parametre ayarlarıyla dengesizliği hesaba kattırarak bir model eğiteceğiz
```

```python
#şimdi korelasyona bakalım label ile sayısal olan değişkenler arasındakine bakalım önce 
corr_cols = cont_cols + ["is_healthy"]
plt.figure(figsize=(5+1.2*len(corr_cols), 4.5))
sns.heatmap(df[corr_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Korelasyon Matrisi (sürekli + is_healthy)")
plt.tight_layout()
plt.savefig(os.path.join(outputdir, "correlation_continuous_ishealthy.png"), bbox_inches="tight", dpi=300)
plt.close()
```

```python
#korelasyonlar oldukca zayıf en bariz negatif fat calori arasında -0.2ye yakın en anlamlı poztif prt o bile 0.06 cok kücük yani
```

```python
#binarylerin korelasyonuna da bakalım
corr_matrix = df[binary_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasyon Matrisi (Binary Değişkenler)')
plt.savefig(os.path.join(outputdir, "correlation_binary_ishealthy.png"), bbox_inches="tight", dpi=300)
plt.close()
```

```python
#bu sefer daha anlamlı oldu özellikle birbirine benzeyen veya zıt olan yemek türleri bariz belli oluyor
```

```python
#is_healthy ye göre sayısal değişkenlerin kutu grafikleri
for c in cont_cols:
    plt.figure(figsize=(6.5, 3.2))
    data0 = df.loc[df["is_healthy"] == 0, c].dropna()
    data1 = df.loc[df["is_healthy"] == 1, c].dropna()

    # labels -> tick_labels
    plt.boxplot([data0, data1], tick_labels=["0", "1"])
    plt.title(f"{c} by is_healthy (0/1)")
    plt.ylabel(c)
    plt.tight_layout()
    plt.savefig(os.path.join(outputdir, f"{c}_by_ishealthy_box.png"),
                bbox_inches="tight", dpi=300)
    plt.close()
```

```python
#yine çok yoğun bir not healthy var
```

```python
# diyet stillerine göre healthy dağılımına bakalım
if binary_cols:
    rate = {}
    for b in binary_cols:
        sub = df[df[b]==1]
        rate[b] = sub["is_healthy"].mean() if len(sub)>0 else np.nan
    rate = pd.Series(rate).sort_values(ascending=False)
    rate.to_csv(os.path.join(outputdir, "healthy_rate_by_flag.csv"))
    plt.figure(figsize=(9, 0.5*len(rate)+2))
    plt.barh(rate.index, (100*rate.values))
    plt.title("Flag=1 iken Sağlıklı Oranı (%)")
    plt.xlabel("percent")
    plt.tight_layout()
    plt.savefig(os.path.join(outputdir, "healthy_rate_by_flag.png"), bbox_inches="tight", dpi=300)
    plt.close()
```
